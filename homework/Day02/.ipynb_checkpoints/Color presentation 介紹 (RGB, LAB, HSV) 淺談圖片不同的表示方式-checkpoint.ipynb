RGB 是非常廣泛的表⽰示⽅方式但是我們在敘述顏⾊色時通常是以甚麼顏⾊色？顏⾊色多鮮豔？顏⾊色多亮？來來定義⼀一個顏⾊色，因此就衍⽣生出以下幾種顏⾊色表⽰示法

HSB/HSV/HSL

Hue 色相  
Saturation 飽和度  
Brightness明度/ Value明度/ Lightness 亮度

HSB = HSV

Hue ⾊相以 360 度決定顏⾊  

Saturation 飽和度以 0~100% 決定⾊彩純度⽩，⿊，灰⾊系⾊彩的飽和度為 0  

Brightness 明度以 0~100% 決定鮮豔程度，黑⾊明度為 0

HSL 在飽和度跟明亮度這邊與 HSB 的表⽰示有點不同  
飽和度  
HSB 的飽和度 0~100% 是⽩色到選擇的色相   

HSL 的飽和度 0~100% 是灰色到選擇的色相，他們認為純白⾊色是飽和的顏色

亮度  
HSB 的明度是光的量，可以是任何顏色e.g. B=100 時看 H, S 決定是甚麼顏色  
HSL 的亮度是⽩色的量e.g. B=100 時是⽩⾊
